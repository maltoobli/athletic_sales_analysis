```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_20_df = pd.read_csv("Resources/athletic_sales_2020.csv")
athletic_21_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
athletic_20_df
```

```python
# Display the 2021 sales DataFrame
athletic_21_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_20_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_21_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales = pd.concat([athletic_20_df, athletic_21_df], ignore_index=True)

combined_sales
```

```python
# Check if any values are null.
combined_sales.isna().any()
```

```python
# Check the data type of each column
combined_sales.count()
```

```python
combined_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales["invoice_date"] = pd.to_datetime(combined_sales["invoice_date"], format="%m/%d/%y")
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
group_product_sold = combined_sales.groupby(["region", "state", "city"])[["units_sold"]].sum()


# Rename the sum to "Total_Products_Sold".
group_product_sold.rename(columns = {"units_sold" : "Total_Products_Sold"}, inplace=True)




# Show the top 5 results.
group_product_sold.sort_values(by="Total_Products_Sold" , ascending=False).head(5)                      
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_product_sold = pd.pivot_table(combined_sales,
                                    values= "units_sold",
                                    index= ["region", "state", "city"],
                                   aggfunc = "sum")

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_product_sold.rename(columns = {"units_sold" : "Total_Products_Sold"}, inplace=True)


# Show the top 5 results.
pivot_product_sold.sort_values(by="Total_Products_Sold" , ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
group_product_total_sales = combined_sales.groupby(["region", "state", "city"])[["total_sales"]].sum()


# Rename the "total_sales" column to "Total Sales"
group_product_total_sales.rename(columns = {"total_sales" : "Total Sales"}, inplace=True)


# Show the top 5 results.
group_product_total_sales.sort_values(by="Total Sales", ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_product_total_sales = pd.pivot_table(combined_sales,
                                    values= "total_sales",
                                    index= ["region", "state", "city"],
                                   aggfunc = "sum")

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_product_total_sales.rename(columns = {"total_sales" : "Total Sales"}, inplace=True)


# Show the top 5 results.
pivot_product_total_sales.sort_values(by="Total Sales", ascending=False).head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
group_total_sales_products = combined_sales.groupby(["retailer", "region", "state", "city"])[["total_sales"]].sum()


# Rename the "total_sales" column to "Total Sales"
group_total_sales_products.rename(columns = {"total_sales" : "Total Sales"}, inplace=True)

# Show the top 5 results.
group_total_sales_products.sort_values(by= "Total Sales", ascending=False)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_total_sales_products= pd.pivot_table(combined_sales,
                                    values= ["total_sales"],
                                    index= ["retailer","region", "state", "city"],
                                   aggfunc = "sum")


# Optional: Rename the "total_sales" column to "Total Sales"
pivot_total_sales_products.rename(columns = {"total_sales" : "Total Sales"}, inplace=True)


# Show the top 5 results.
pivot_total_sales_products.sort_values(by= "Total Sales", ascending=False)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_athletic_footwear = combined_sales[combined_sales["product"].str.contains("Women's Athletic Footwear")]
women_athletic_footwear

#onther option:
#combined_sales.loc[(combined_sales["product"] =="Women's Athletic Footwear")].
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
group_total_women_athletic_footwear_products = women_athletic_footwear.groupby(["retailer", "region", "state", "city"])[["units_sold"]].sum()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
group_total_women_athletic_footwear_products.rename(columns = {"units_sold" : "Womens_Footwear_Units_Sold"}, inplace=True)

# Show the top 5 results.
group_total_women_athletic_footwear_products.sort_values(by= "Womens_Footwear_Units_Sold", ascending=False)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

pivot_total_women_athletic_footwear_products= pd.pivot_table(women_athletic_footwear,
                                    values= ["units_sold"],
                                    index= ["retailer","region", "state", "city"],
                                   aggfunc = "sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

pivot_total_women_athletic_footwear_products.rename(columns = {"units_sold" : "Womens_Footwear_Units_Sold"}, inplace=True)

# Show the top 5 results.
pivot_total_women_athletic_footwear_products.sort_values(by= "Womens_Footwear_Units_Sold", ascending=False)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pivot_total_women_athletic_footwear_sales= pd.pivot_table(women_athletic_footwear,
                                    values= ["total_sales"],
                                    index= ["invoice_date"],)

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_total_women_athletic_footwear_sales.rename(columns={"total_sales" : "Total Sales"}, inplace=True)


# Show the table.
pivot_total_women_athletic_footwear_sales
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

daily_sales = pivot_total_women_athletic_footwear_sales.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales_sorted = daily_sales.sort_values(by='Total Sales', ascending=False).head(10)
daily_sales_sorted
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pivot_total_women_athletic_footwear_sales.resample('W').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales_sorted = weekly_sales.sort_values(by='Total Sales', ascending=False)
weekly_sales_sorted.head(10)
```
